```python
import pandas as pd
import requests
```

## Using API

```python
response = requests.get('https://api.themoviedb.org/3/movie/top_rated?api_key=90de64ccd37cdb5f60e7b03015b62dae&language=en-US&page=1') # Use API key and run
```

```python
response.json()['results'][0:4]
```

Since this is in a dict format, we can convert it to a dataframe to use it for easier computation and select the required columns

```python
temp_df=pd.DataFrame(response.json()['results'])[['id','title','overview','release_date','popularity','vote_average','vote_count']]
```

```python
temp_df
```

### The above code is working fine but it is for page 1 out of 428 pages. So we need to extract all the pages from the api by adding the request in a for loop.

```python
final_df = pd.DataFrame()  # Create an empty dataframe to store the data
for i in range(1,429):
  response = requests.get('https://api.themoviedb.org/3/movie/top_rated?api_key=90de64ccd37cdb5f60e7b03015b62dae&language=en-US&page={pagenum}'.format(pagenum=i))
  temp_df = pd.DataFrame(response.json()['results'])[['id','title','overview','release_date','popularity','vote_average','vote_count']]
  final_df = final_df._append(temp_df,ignore_index=True)
```
