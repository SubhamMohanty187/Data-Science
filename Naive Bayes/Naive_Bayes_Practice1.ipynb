```python
wine_df = pd.read_csv("wine.csv", names = ["Cultivator", "Alchol", "Malic_Acid", "Ash", "Alcalinity_of_Ash", "Magnesium", "Total_phenols", "Falvanoids", "Nonflavanoid_phenols", "Proanthocyanins", "Color_intensity", "Hue", "OD280", "Proline"])
df = wine_df.copy()
df.head(10)
```

```python
df.shape
```

```python
df["Cultivator"].value_counts() 
```

```python
X = df.drop("Cultivator" , axis = 1)
y = df["Cultivator"].values
```

```python
sns.pairplot(df,hue="Cultivator",diag_kind="kde")
plt.show()
```

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(X,y,test_size=0.30,random_state=1)
```

```python
model.fit(x_train,y_train)
```

```python
y_predict = model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
acc = accuracy_score(y_test,y_predict)
print(acc*100,"%")
```
