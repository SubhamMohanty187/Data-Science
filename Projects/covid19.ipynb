```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# New Section

```python
covid=pd.read_csv('/content/Latest Covid-19 India Status.csv')
covid.head()
```

```python
covid.shape
```

```python
covid.isna().sum()
```

```python
cv=covid.copy()
cv.head()
```

```python
cv=cv.sort_values(by=['Total Cases'],ascending=False)
cv.reset_index(drop=True,inplace=True)
```

```python
sns.set()
plt.figure(figsize=(10,5))
plt.title('STATE vs TOTAL CASES')
plt.bar(cv['State/UTs'],cv['Total Cases'],color='Red')
plt.xticks(rotation=90)
plt.xlabel('States/UTs')
plt.ylabel('Total Cases In Million')
#plt.grid(True)

plt.show()
```

```python
#sns.set()
#sns.set(rc={'figure.figsize':(100,20)})
plt.figure(figsize=(10,5))
plt.plot(cv['State/UTs'],cv['Deaths'],label='Deaths',marker="o")
plt.title("Death Cases per State")
plt.xticks(rotation=90)
plt.legend()
plt.autoscale

plt.show()
```

```python
cv=cv.sort_values(by=['Active'],ascending=False)
cv.reset_index(drop=True,inplace=True)
cv.head()
```

```python
plt.pie(cv["Active"].loc[0:5],labels=cv["State/UTs"].loc[0:5], autopct='%.1f%%',startangle=90)
plt.axis("equal")
plt.title("Top 5 states with highest active cases.")
plt.show()
```

```python
plt.pie(cv["Deaths"].loc[0:5],labels=cv["State/UTs"].loc[0:5], autopct='%.1f%%',startangle=40)
plt.axis("equal")
plt.title("Top 5 states with highest deaths.")
plt.show()
```

```python
cv=cv.sort_values(by=['Discharged'],ascending=False)
cv.reset_index(drop=True,inplace=True)
cv.head()
```

```python
plt.figure(figsize=(10,5))
plt.title('Discharge rates across States')
plt.bar(cv['State/UTs'],cv['Discharged'],color='green')
plt.xticks(rotation=90)
plt.xlabel('States/UTs')
plt.ylabel('Total Cases In Million')
```
