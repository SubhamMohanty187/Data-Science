### Exploratory Data Analysis on IRIS datset

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.get_dataset_names()
```

```python
df = sns.load_dataset("iris")
df.head()
```

```python
print(df.shape)  #dataset has 150 rows and 5 columns.
```

```python
df.columns  # Shows all the columns in our dataset and species is class-label.
```

###If the no. of datapoints is equal for each class then it is called balanced dataset else it called imbalanced dataset. Thus, iris is balanced dataset.

```python
df['species'].value_counts()  #how many data-points are present for each class of class-label.
```

Here, hue adds different color to different species and also adds a legend for better understanding. The parameter "s" will set the size of dots in the scatterplot.

```python
sns.set()
sns.relplot(x="sepal_length",y="sepal_width",hue="species",s=80, data =df,kind="scatter")   
plt.show()
```

##Pair Plots
Here, a matrix 4x4 size of four parameters sl,sw,pl,pw. and the daigonals are PDF for each feature.
If the dimensionality or features is more, pair plots are not useful.

```python
sns.pairplot(df, hue="species")
plt.show()
```

```python
sns.FacetGrid(df, hue="species", size=5) \
   .map(sns.distplot, "petal_length") \
   .add_legend();
plt.show();
```

```python
sns.histplot(data=df,x=df["petal_length"],hue="species")
counts, bin_edges = np.histogram(df['petal_length'], bins=10,  density = True)
pdf = counts/(sum(counts))
print(pdf);
plt.show()
```

Boxplot or Whiskerplot

```python
sns.boxplot(x="species",y="petal_length",data=df)
plt.show()
```

###Violin plots

```python
sns.violinplot(x="species",y="petal_length",data=df,size=10)
plt.show()
```

##Multivariate probabilty density or Joint Plot

####For setosa species only

```python
df=df.loc[df["species"]=="setosa"]
sns.set_style("white")
sns.jointplot(x="petal_length", y="petal_width", data=df, kind="kde",cmap="Blues",shade=True)  #cmap =s contourmap and shade represents density
plt.show();
```
